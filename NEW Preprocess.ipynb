```python
def matplot_plot(frame,name_sensor_plot):
    plt.figure(figsize=(14, 7))
    headers = imu.find_headers(frame)
    plt.title(f"{name_sensor_plot}")
    plt.plot(frame.index, frame[headers[3]], 'b', alpha=0.75)
    plt.plot(frame.index, frame[headers[4]], 'r',frame.index,  frame[headers[5]], 'g')
#     plt.legend(str(headers[0]), str(headers[0]), str(headers[0]), loc='lower right')
    plt.grid(True)
```

```python
matplot_plot(bt,'bt non- calibrated')
matplot_plot(calibrated_bt,' bt Calibrated')
```

```python
matplot_plot(bb,'bb non- calibrated')
matplot_plot(calibrated_bb,' bb Calibrated')
```

```python
matplot_plot(lt,'lt non- calibrated')
matplot_plot(calibrated_lt,' lt Calibrated')
```
